```python
import os
from PIL import Image

DATASET_PATH = r"E:\Kidney project\Kidney-disease-classification-with-mlflow-dvc\artifacts\data_ingestion\data\Kidney-CT-Scan-Images"

def test_image_access():
    if not os.path.exists(DATASET_PATH):
        print("‚ùå Path does NOT exist:", DATASET_PATH)
        return

    print("‚úÖ Folder exists!")

    classes = os.listdir(DATASET_PATH)
    print("‚úÖ Classes found:", classes)

    for cls in classes:
        class_path = os.path.join(DATASET_PATH, cls)

        if not os.path.isdir(class_path):
            continue

        print(f"\nüîç Checking class: {cls}")
        images = os.listdir(class_path)

        if len(images) == 0:
            print("   ‚ùå No images found!")
            continue

        # Test first image
        first_image = os.path.join(class_path, images[0])

        try:
            img = Image.open(first_image)
            img.verify()  # check corruption
            print("   ‚úÖ First image opened successfully:", first_image)
        except Exception as e:
            print("   ‚ùå Cannot open image:", first_image)
            print("   Error:", e)

test_image_access()
```
